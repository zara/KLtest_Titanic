```python
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.impute import SimpleImputer
from scipy.stats import poisson
sns.set()
```

```python
def kl_divergence(p, q):
    return np.sum(np.where(np.logical_and(p != 0, q != 0), p * np.log(p / q), 0))
```

```python
df = pd.read_csv("titanic.csv")
df['Age'].fillna(df['Age'].median(), inplace=True)
females_age=df['Age'][df['Sex'].isin(["female"])]
count, devision = np.histogram(females_age, bins=200)
total_counts = len(females_age)

data_norm = poisson.rvs(mu=27.9299, size=total_counts)
count2, devision2 = np.histogram(data_norm, bins=200)
```

```python
plt.title('KL')
print(kl_divergence(count, count2))
plt.plot(count, c='blue')
plt.plot(count2, c='red')
```
